```python
root = "./../"

entities = set()
relations = set()
classes = set()

for name in ["FD", "TR"]:
	for granularity in ["Y", "D"]:
		for size in ["Extra_Small", "Small", "Medium", "Large"]:
			sub_folder = f"Data/{name}_{granularity}/{size}/"


			with open(f"{root}{sub_folder}/Entity2ID.txt", "r", encoding="UTF-8") as f_r:
				line = f_r.readline()
				while line!="":
					ent, _ = line.split("\t")
					entities.add(ent)
					line = f_r.readline()
				
			with open(f"{root}{sub_folder}/Relation2ID.txt", "r", encoding="UTF-8") as f_r:
				line = f_r.readline()
				while line!="":
					rel, _ = line.split("\t")
					relations.add(rel)
					line = f_r.readline()
				
			with open(f"{root}{sub_folder}/Class2ID.txt", "r", encoding="UTF-8") as f_r:
				line = f_r.readline()
				while line!="":
					rel, _ = line.split("\t")
					classes.add(rel)
					line = f_r.readline()

	# Query

	## Natural Language

with open(f"{root}Queries/query_nl_entities.q", "w", encoding="UTF-8") as f_w:
	f_w.write("""
PREFIX wd: <http://www.wikidata.org/entity/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX schema: <http://schema.org/>
		
SELECT ?iri ?iriLabel ?iriDescription
WHERE {
VALUES ?iri {wd:""")
	f_w.write(" wd:".join(entities))
	f_w.write("""
}
?iri rdfs:label ?iriLabel.
FILTER(lang(?iriLabel) = "en")

OPTIONAL{
	?iri schema:description ?iriDescription.
	FILTER(lang(?iriDescription) = "en")
}
}
""")


with open(f"{root}Queries/query_nl_relations.q", "w", encoding="UTF-8") as f_w:
	f_w.write("""
PREFIX wd: <http://www.wikidata.org/entity/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX schema: <http://schema.org/>
		
SELECT ?iri ?iriLabel ?iriDescription
WHERE {
VALUES ?iri {wd:""")
	f_w.write(" wd:".join(relations))
	f_w.write("""
}
?iri rdfs:label ?iriLabel.
FILTER(lang(?iriLabel) = "en")

OPTIONAL{
	?iri schema:description ?iriDescription.
	FILTER(lang(?iriDescription) = "en")
}
}
""")
	
with open(f"{root}Queries/query_nl_classes.q", "w", encoding="UTF-8") as f_w:
	f_w.write("""
PREFIX wd: <http://www.wikidata.org/entity/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX schema: <http://schema.org/>
		
SELECT ?iri ?iriLabel ?iriDescription
WHERE {
VALUES ?iri {wd:""")
	f_w.write(" wd:".join(classes))
	f_w.write("""
}
?iri rdfs:label ?iriLabel.
FILTER(lang(?iriLabel) = "en")

OPTIONAL{
	?iri schema:description ?iriDescription.
	FILTER(lang(?iriDescription) = "en")
}
}
""")

### Static Information

#		with open(f"{root}Queries/query_static_entities_{name}_{limit}.q", "w", encoding="UTF-8") as f_w:
#			f_w.write("""
#PREFIX wd: <http://www.wikidata.org/entity/>
#PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
#PREFIX schema: <http://schema.org/>
		   
#SELECT ?iri ?r ?v
#WHERE {
#	VALUES ?iri {wd:""")
#			f_w.write(" wd:".join(list(entities)))
#			f_w.write("""
#	}
#	?iri ?r ?v. 
#	FILTER(strstarts(str(?r), "http://www.wikidata.org/prop/direct/"))
#}
#""")
```
