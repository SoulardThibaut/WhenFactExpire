```python
import numpy as np
import os 

names = ["Large", "Medium", "Small", "Extra_Small"]
types_data = ["TR", "FD"]
temporal_precisions = ["Y", "D"]
```

```python
def create_folder(directory:str):
	if not os.path.exists(directory):
		os.makedirs(directory)

def fill_in_the_gaps(start:np.datetime64, end:np.datetime64):
	timestamps_in_between = [start]
	gap = start 
	while gap != end :
		gap+=1 
		timestamps_in_between.append(gap)
	return timestamps_in_between

def write_commands(f_w, folder, temporal_precision):
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TimePlex_base -a '{"embedding_dim":100, "srt_wt":5.0, "ort_wt":5.0, "sot_wt":5.0, "emb_reg_wt":0.005}' -l crossentropy_loss_AllNeg -r 0.1 -b 500 -x -1 -n 0 -v 1 -q 0 -y 10 -g_reg 2 -g 2.0 --perturb_time 1 --mode train --flag_add_reverse 1 -e 100 --save_dir """+f"{folder}"+"""_TimePlex_base --timeout 7200 --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TimePlex_base --resume_from_save "./models/"""+f"{folder}"+"""_TimePlex_base/best_valid_model.pt" --mode test --predict_time 1 -y 10 --save_time_results "./Res_server/"""+f"{folder}"+"""/TimePlex_base" --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TimePlex -a '{"embedding_dim":100, "model_path":"./models/"""+f"{folder}"+"""_TimePlex_base/best_valid_model.pt", "pairs_wt":3.0, "pairs_args":{"reg_wt":0.002}}' -l crossentropy_loss -r 0.05 -b 500 -x -1 -n 100 -v 1 -q 0 -y 10  -g 1.0 -bt 0 --patience 1 -e 20 --save_dir """+f"{folder}"+"""_TimePlex --timeout 7200 --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TimePlex --resume_from_save "./models/"""+f"{folder}"+"""_TimePlex/best_valid_model.pt"  --mode test --predict_time 1 -y 10 --save_time_results "./Res_server/"""+f"{folder}"+"""/TimePlex" -bt 0 --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TComplex_lx -a '{"embedding_dim":100}' -l crossentropy_loss -r 0.05 -b 500 -x -1 -n 100 -v 1 -q 0 -y 10  -g 1.0 --patience 1 -e 100 --save_dir """+f"{folder}"+"""_TComplex_lx --timeout 7200 --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TComplex_lx --resume_from_save "./models/"""+f"{folder}"+"""_TComplex_lx/best_valid_model.pt"  --mode test --predict_time 1 -y 10 --save_time_results "./Res_server/"""+f"{folder}"+"""/TComplex_lx" --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TNTComplex_lx -a '{"embedding_dim":100}' -l crossentropy_loss -r 0.05 -b 500 -x -1 -n 100 -v 1 -q 0 -y 10  -g 1.0 --patience 1 -e 100 --save_dir """+f"{folder}"+"""_TNTComplex_lx --timeout 7200 --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TNTComplex_lx --resume_from_save "./models/"""+f"{folder}"+"""_TNTComplex_lx/best_valid_model.pt"  --mode test --predict_time 1 -y 10 --save_time_results "./Res_server/"""+f"{folder}"+"""/TNTComplex_lx" --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("\n\n")

def write_commands_train(f_w, folder,  temporal_precision):
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TimePlex_base -a '{"embedding_dim":100, "srt_wt":5.0, "ort_wt":5.0, "sot_wt":5.0, "emb_reg_wt":0.005}' -l crossentropy_loss_AllNeg -r 0.1 -b 500 -x -1 -n 0 -v 1 -q 0 -y 10 -g_reg 2 -g 2.0 --perturb_time 1 --mode train --flag_add_reverse 1 -e 100 --save_dir """+f"{folder}"+"""_TimePlex_base --timeout 7200 --time_precision """+f'"{temporal_precision}"'+""" \n""")
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TimePlex -a '{"embedding_dim":100, "model_path":"./models/"""+f"{folder}"+"""_TimePlex_base/best_valid_model.pt", "pairs_wt":3.0, "pairs_args":{"reg_wt":0.002}}' -l crossentropy_loss -r 0.05 -b 500 -x -1 -n 100 -v 1 -q 0 -y 10  -g 1.0 -bt 0 --patience 1 -e 20 --save_dir """+f"{folder}"+"""_TimePlex --timeout 7200 --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TComplex_lx -a '{"embedding_dim":100}' -l crossentropy_loss -r 0.05 -b 500 -x -1 -n 100 -v 1 -q 0 -y 10  -g 1.0 --patience 1 -e 100 --save_dir """+f"{folder}"+"""_TComplex_lx --timeout 7200 --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TNTComplex_lx -a '{"embedding_dim":100}' -l crossentropy_loss -r 0.05 -b 500 -x -1 -n 100 -v 1 -q 0 -y 10  -g 1.0 --patience 1 -e 100 --save_dir """+f"{folder}"+"""_TNTComplex_lx --timeout 7200 --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("\n\n")

def write_commands_validation(f_w, folder,  temporal_precision):
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TimePlex_base --resume_from_save "./models/"""+f"{folder}"+"""_TimePlex_base/best_valid_model.pt" --mode test --predict_time 1 -y 10 --save_time_results "./Res_server/"""+f"{folder}"+"""/TimePlex_base" --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TimePlex --resume_from_save "./models/"""+f"{folder}"+"""_TimePlex/best_valid_model.pt"  --mode test --predict_time 1 -y 10 --save_time_results "./Res_server/"""+f"{folder}"+"""/TimePlex" -bt 0 --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TComplex_lx --resume_from_save "./models/"""+f"{folder}"+"""_TComplex_lx/best_valid_model.pt"  --mode test --predict_time 1 -y 10 --save_time_results "./Res_server/"""+f"{folder}"+"""/TComplex_lx" --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("""python3 -u main.py -d """+f"{folder}"+""" -m TNTComplex_lx --resume_from_save "./models/"""+f"{folder}"+"""_TNTComplex_lx/best_valid_model.pt"  --mode test --predict_time 1 -y 10 --save_time_results "./Res_server/"""+f"{folder}"+"""/TNTComplex_lx" --time_precision """+f'"{temporal_precision}"'+"""\n""")
	f_w.write("\n\n")
```

```python
for type_data in types_data:
	for temporal_precision in temporal_precisions:
		for name in names:
			
			root_data_input = f"./../Retrieve_Data/Data/{type_data}_{temporal_precision}/{name}/"
			root_data_ouput = f"./data/{type_data}_{temporal_precision}_{name}/" 

			Unk_s = None
			if type_data == "TR":
				Unk_s = np.datetime64("1900-01-01", temporal_precision)
			else:
				Unk_s = np.datetime64("-1000-01-01", temporal_precision)

			Unk_e = None

			with open(f"{root_data_input}OP_train.txt", "r", encoding="UTF-8") as f_r:
				line = f_r.readline()
				while line != "":
					if len(line[:-1].split("\t")) == 5:
						h,r,v,st,en = line[:-1].split("\t")
						
						if st != "None":
							if (not Unk_s )or  (Unk_s > np.datetime64(st, temporal_precision)):
								Unk_s = np.datetime64(st, temporal_precision) 

							if (not Unk_e )or  (Unk_e < np.datetime64(st, temporal_precision)):
								Unk_e = np.datetime64(st, temporal_precision)+1

						if en != "None":
							if (not Unk_e )or (Unk_e < np.datetime64(en, temporal_precision)):
								Unk_e = np.datetime64(en, temporal_precision)

							if (not Unk_s )or  (Unk_s > np.datetime64(en, temporal_precision)):
								Unk_s = np.datetime64(en, temporal_precision)-1

						
					line = f_r.readline()
				print( Unk_s, Unk_e)

				t_seen = set()
			create_folder(f"{root_data_ouput}")
			create_folder(f"{root_data_ouput}ValidationTask")
			create_folder(f"{root_data_ouput}intervals")

			for (f, f_gt) in [("OP_train.txt", "OP_train.gt"), ("OP_test.txt", "OP_test.gt"), ("OP_valid.txt", "OP_valid.gt")]:
				with open(f"{root_data_input}{f}", "r", encoding="UTF-8") as f_read:
					with open(f"{root_data_input}{f_gt}", "r", encoding="UTF-8") as f_read_gt:
						with open(f"{root_data_ouput}ValidationTask/{f.split('.')[0].split("_")[1]}.txt", "w", encoding="UTF-8") as f_out:
							with open(f"{root_data_ouput}ValidationTask/{f.split('.')[0].split("_")[1]}.gt", "w", encoding="UTF-8") as f_out_gt:
								line = f_read.readline()
								line_gt = f_read_gt.readline()
								while line != "":
									if len(line[:-1].split("\t")) == 5:
										h,r,v,st,en = line[:-1].split("\t")
										if st == "None":
											st = Unk_s
										else:
											st = np.datetime64(st, temporal_precision)
										
										if en == "None":
											en = Unk_e
										else:
											en = np.datetime64(en, temporal_precision)
										
										if st <= en:
											if f == "OP_train.txt":
												if line_gt[:-1] == "True":
													t_seen.add(st)
													t_seen.add(en)
												f_out.write(f"{h}\t{r}\t{v}\t{st}\t{en}.\n")
												f_out_gt.write(line_gt)
											else:
												#if (st in t_seen) and (en in t_seen):
												f_out.write(f"{h}\t{r}\t{v}\t{st}\t{en}.\n")
												f_out_gt.write(line_gt)
										else:
											print(line)
									else:
										h, r, v, t = line[:-1].split("\t")
										t_np_s = np.datetime64(t, temporal_precision)
										t_np_e = t_np_s+1
										if f == "OP_train.txt":
											if line_gt[:-1] == "True":
												t_seen.add(t_np_s)
												t_seen.add(t_np_e)
											f_out.write(f"{h}\t{r}\t{v}\t{t_np_s}\t{t_np_e}.\n")
											f_out_gt.write(line_gt)
										else:
											#if (t_np_s in t_seen) and (t_np_e in t_seen):
											f_out.write(f"{h}\t{r}\t{v}\t{t_np_s}\t{t_np_e}.\n")
											f_out_gt.write(line_gt)

									line = f_read.readline()
									line_gt = f_read_gt.readline()


			for f in ["train", "valid", "test"]:
				f_r_d = open(f"{root_data_ouput}ValidationTask/{f}.txt", "r", encoding="UTF-8")
				f_r_gt = open(f"{root_data_ouput}ValidationTask/{f}.gt", "r", encoding="UTF-8")
				f_w_d = open(f"{root_data_ouput}{f}.txt", "w", encoding="UTF-8")

				line_d = f_r_d.readline()
				line_gt = f_r_gt.readline()

				while line_d!="":

					if line_gt[:-1] == "True":
						f_w_d.write(line_d)

					line_d = f_r_d.readline()
					line_gt = f_r_gt.readline()
				
				f_r_d.close()
				f_r_gt.close()
				f_w_d.close()

				if f != "train":
					os.system(f"cp {root_data_ouput}{f}.txt {root_data_ouput}intervals/{f}.txt")

			with open(f"{root_data_ouput}all_timestamps.txt", "w", encoding="UTF-8") as f_w:
				start, end = Unk_s, Unk_e
				gaps = fill_in_the_gaps(start, end+1) # In case a timestamp happens at the last timestamp allowed thus forcing a +1
				for gap in gaps:
					f_w.write(f"{gap}\n")

			create_folder("./Scripts_server/")

			folders = [root_data_ouput.split("/")[-2]]
			for folder in folders:
					with open(f"./Scripts_server/{folder}.slurm", "w", encoding="UTF-8") as f_w: 
						f_w.write("""#!/bin/bash
#SBATCH --gres=gpu:1\n\n
# cd ..\n""")
#						f_w.write("""mkdir ./Monitoring/$(date +%s)

## Define log files
#CPU_LOG="./Monitoring/$(date +%s)/cpu.log"
#GPU_LOG="./Monitoring/$(date +%s)/gpu.log"

## Start CPU/RAM monitoring (1-second interval)
#vmstat 1 > $CPU_LOG &
#VMSTAT_PID=$!
			
## Start GPU monitoring (1-second interval)
#while true; do nvidia-smi --query-gpu=name,timestamp,utilization.gpu,utilization.memory,memory.used,memory.total --format=csv >> $GPU_LOG; sleep 25; done &
#NVIDIA_MONITOR_PID=$!
#""")
						write_commands(f_w, folder, temporal_precision)
#						f_w.write("""# Clean up monitoring processes
#kill $VMSTAT_PID
#kill $NVIDIA_MONITOR_PID""")
					
#			for folder in [ f"{root_data_ouput.split("/")[-2]}"]:
#				with open(f"./Scripts_server/{folder}_train.slurm", "w", encoding="UTF-8") as f_w: 
#					f_w.write("""#!/bin/bash
##SBATCH --gres=gpu:1\n\n
## cd ..\n""")
#					f_w.write("""mkdir ./Monitoring/$(date +%s)

## Define log files
#CPU_LOG="./Monitoring/$(date +%s)/cpu.log"
#GPU_LOG="./Monitoring/$(date +%s)/gpu.log"

## Start CPU/RAM monitoring (1-second interval)
#vmstat 1 > $CPU_LOG &
#VMSTAT_PID=$!
			
## Start GPU monitoring (1-second interval)
#while true; do nvidia-smi --query-gpu=name,timestamp,utilization.gpu,utilization.memory,memory.used,memory.total --format=csv >> $GPU_LOG; sleep 25; done &
#NVIDIA_MONITOR_PID=$!
#""")
#					write_commands_train(f_w, folder, temporal_precision)
#					f_w.write("""# Clean up monitoring processes
#kill $VMSTAT_PID
#kill $NVIDIA_MONITOR_PID""")
			
			
#			for folder in [ f"{root_data_ouput.split("/")[-2]}"]:
#				with open(f"./Scripts_server/{folder}_valid.slurm", "w", encoding="UTF-8") as f_w: 
#					f_w.write("""#!/bin/bash
##SBATCH --gres=gpu:1\n\n
## cd ..\n""")
#					f_w.write("""mkdir ./Monitoring/$(date +%s)

## Define log files
#CPU_LOG="./Monitoring/$(date +%s)/cpu.log"
#GPU_LOG="./Monitoring/$(date +%s)/gpu.log"

## Start CPU/RAM monitoring (1-second interval)
#vmstat 1 > $CPU_LOG &
#VMSTAT_PID=$!
			
## Start GPU monitoring (1-second interval)
#while true; do nvidia-smi --query-gpu=name,timestamp,utilization.gpu,utilization.memory,memory.used,memory.total --format=csv >> $GPU_LOG; sleep 25; done &
#NVIDIA_MONITOR_PID=$!
#""")
#					write_commands_validation(f_w, folder, temporal_precision)
#					f_w.write("""# Clean up monitoring processes
#	kill $VMSTAT_PID
#	kill $NVIDIA_MONITOR_PID""")

			create_folder("Res_server")
			for folder in folders:
				create_folder(f"./Res_server/{folder}/")
					

```

```python
for name in names:
	with open(f"./Scripts_server/{name}.slurm", "w", encoding="UTF-8") as f_w: 
		f_w.write("""#!/bin/bash
#SBATCH --gres=gpu:1\n\n""")
		
		for type_data in types_data:
			for temporal_precision in temporal_precisions:
				folder = f"{type_data}_{temporal_precision}_{name}"
				write_commands(f_w, folder, temporal_precision)
```
